```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('dark_background')
```

```python
np.random.seed(0)
data0= np.random.randn(1000)
```

```python
plt.hist(data0, bins=30, color='orange', edgecolor='gray')
plt.title('Basic Histogram')
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.hist(data0, bins=100, density=True, color='orange', edgecolor='gray')
plt.title('Basic Histogram')
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.hist(data0, bins=30, density=True, color='orange',linestyle='--', edgecolor='gray')
plt.title('Basic Histogram')
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.style.use('fast')
data1= np.random.randn(1000)
data2= np.random.randn(1000) + 2

plt.hist(data1, bins=30, alpha=0.5, label='Data 1', color='blue', edgecolor='white')
plt.hist(data2, bins=30, alpha=0.5, label='Data 2', color='red', edgecolor='white')

plt.title=('Superpostion of Multiple histograms')
plt.xlabel=('Values')
plt.ylabel=('Frequency')
plt.legend()
plt.show()
```

```python
import seaborn as sns
from importlib import reload
plt=reload(plt)

sns.histplot(data0, bins=30, kde=True, color='green')
plt.suptitle('KDE Density Kernel Histogram')
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.show()
```

```python
fig, axs = plt.subplots(1, 2, figsize=(12, 6))

axs[0].hist(data1, bins=30, color='blue', edgecolor='black')
axs[0].set_title('Histogram 1')
axs[0].set_xlabel('Values')
axs[0].set_ylabel('Frequency')

axs[1].hist(data2, bins=30, color='red', edgecolor='black')
axs[1].set_title('Histogram 2')
axs[1].set_xlabel('Values')
axs[1].set_ylabel('Frequency')

plt.tight_layout()
plt.show
```

```python
plt.hist(data0, bins=30, color='orange', edgecolor='gray')
plt.suptitle('Basic Histogram')
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.hist(data0, bins=30, color='purple', edgecolor='gray', alpha=0.7)
plt.axvline(data0.mean(), color='red', linestyle='dashed', linewidth=1)
min_ylim, max_ylim = plt.ylim()
plt.text(data0.mean()*1.1, max_ylim*0.9, 'Media: {:.2f}'.format(data0.mean()), color='red')
plt.suptitle('Advanced  Histogram')
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
